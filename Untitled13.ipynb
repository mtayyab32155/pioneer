```python
netf['Release_Date']=pd.to_datetime(netf['Release_Date'])
print(netf['Release_Date'].dtype)
```

**Colums that are not relevant in our analysis are being dropped**

```python
# Drop columns that are not needed for the requested analysis
cols = ['Overview', 'Original_Language', 'Poster_Url']
netf.drop(cols, axis=1, inplace=True)
netf.columns
```

**Now we will categorize the "Vote_Average" column into four categories: verylow, low, medium, high**

```python
def categorize_col(netf, col, label):
  desc_stats = netf[col].describe()
  edges=[desc_stats['min'], desc_stats['25%'], desc_stats['50%'], desc_stats['75%'], desc_stats['max']]
  netf[col]=pd.cut(netf[col], edges, labels=label, duplicates= 'drop')
  return netf
netf.head()
```

```python


labels=['verylow', 'low', 'medium', 'high']

# Apply categorization only to 'Vote_Average'
netf = categorize_col(netf, 'Vote_Average', labels)

print("Vote_Average unique categories after categorization:")
print(netf['Vote_Average'].unique())
print("Popularity dtype after changes:")
print(netf['Popularity'].dtype)

netf.head()
```
